```python
# Remove constant columns as these columns contain the same value for every employee.
df.drop(
    ['EmployeeCount','Over18','StandardHours'],
    axis=1,
    inplace=True
)
```

```python
# Stage 3: Attrition Overview
# Finding how many employees left the company.
```

```python
df['Attrition'].value_counts()
```

```python
df['Attrition'].value_counts().plot(
    kind='bar'
)

plt.title("Employee Attrition")
plt.show()
```

```python
# Stage 4: Attrition by Department
# Finding which department experiences the highest employee turnover.
```

```python
pd.crosstab(
    df['Department'],
    df['Attrition']
)
```

```python
sns.countplot(
    x='Department',
    hue='Attrition',
    data=df
)

plt.xticks(rotation=20)
plt.show()
```

```python
# Stage 5: Attrition by Overtime
# Checking whether overtime affects employee retention.
```

```python
pd.crosstab(
    df['OverTime'],
    df['Attrition']
)
```

```python
sns.countplot(
    x='OverTime',
    hue='Attrition',
    data=df
)

plt.show()
```

```python
# Stage 6: Attrition by Job Satisfaction
# Analyzing the relationship between job satisfaction and attrition.
```

```python
pd.crosstab(
    df['JobSatisfaction'],
    df['Attrition']
)
```

```python
sns.countplot(
    x='JobSatisfaction',
    hue='Attrition',
    data=df
)

plt.show()
```

```python
# Stage 7: Attrition by Monthly Income
# Understanding whether salary influences attrition.
```

```python
plt.figure(figsize=(8,5))

sns.boxplot(
    x='Attrition',
    y='MonthlyIncome',
    data=df
)

plt.show()
```

```python
# Stage 8: Attrition by Work-Life Balance
# Examining how work-life balance affects employee retention.
```

```python
pd.crosstab(
    df['WorkLifeBalance'],
    df['Attrition']
)
```

```python
sns.countplot(
    x='WorkLifeBalance',
    hue='Attrition',
    data=df
)

plt.show()
```

```python
# Key Findings :
# 1. Most employees stayed with the company, but a noticeable number left, showing that employee attrition is still an important issue.
# 2. Employees who worked overtime were more likely to leave the company compared to those who did not work overtime.
# 3. Lower job satisfaction was linked with higher attrition, suggesting that employee satisfaction plays an important role in retention.
# 4. Employees with lower monthly income showed a higher tendency to leave the organization.
# 5. Better work-life balance appeared to reduce attrition, while employees with poor work-life balance were more likely to leave.
```

```python
# Recommendations :
# 1. Reduce excessive overtime and distribute work more effectively among employees.
# 2. Conduct regular feedback sessions to understand employee concerns and improve job satisfaction.
# 3. Review salary structures and provide fair growth opportunities for employees.
# 4. Encourage a healthy work-life balance through flexible work policies and employee wellness initiatives.
# 5. Focus more on departments with higher attrition and take proactive steps to improve employee retention.
```
